```python
#Step1 - load the dataset
import pandas as pd
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
movies.head()
```

```python
#step2 - Merge datasets
movies = movies.merge(credits, on="title")
movies.head()
```

```python
#step3 - Select Important Columns
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
#step4 - Check Missing Data
movies.isnull().sum()
```

```python
#step5 - Handle Missing Values
movies.dropna(inplace=True)
```

```python
movies.iloc[0]['genres']
```

```python
movies.iloc[0]
```

```python
#step6 - Convert JSON like Data to Clean lists
#step6.1 - Import ast
import ast
```

```python
#step6.2 - Create Convert Function
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
#Step6.3 - Apply to Columns
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
#step6.4 - Handle Cast(Top 3 only)
def convert3(text):
    L = []
    count = 0
    for i in ast.literal_eval(text):
        if count < 3:
            L.append(i['name'])
            count += 1
        else: 
            break
    return L
movies['cast'] = movies['cast'].apply(convert3)
    
```

```python
#step6.5 - Extract Director from Crew
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
        return L
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.iloc[0]
```

```python
#Ensure Everything is List
movies['overview'] = movies['overview'].apply(lambda x: x if isinstance(x, list) else [])
movies['genres'] = movies['genres'].apply(lambda x: x if isinstance(x, list) else [])
movies['keywords'] = movies['keywords'].apply(lambda x: x if isinstance(x, list) else [])
movies['cast'] = movies['cast'].apply(lambda x: x if isinstance(x, list) else [])
movies['crew'] = movies['crew'].apply(lambda x: x if isinstance(x, list) else [])
```

```python
#Step7.1 — Combine Columns
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
#step7.2 - create new clean dataframe
new_df = movies[['movie_id','title','tags']]
```

```python
#step7.3 - convert list to string
new_df.loc[:, 'tags'] = new_df['tags'].apply(lambda x: " ".join(x) if isinstance(x, list) else "")
```

```python
#step7.4 - lowercase
new_df.loc[:, 'tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df.head()
```

```python
#step8 - Convert Text to numbers
#step8.1 - import countvecotrizer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#step8.2 - Apply Vecotrizaton
cv = CountVectorizer(max_features=5000,stop_words="english",)
```

```python
#step8.3 - Transfom Data
vectors = cv.fit_transform(new_df["tags"]).toarray()
```

```python
#step8.4 - Check Shape
vectors.shape
```

```python
#step9 - Count Similarity
#step9.1 - import Cosine Similarity
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#step9.2 - Calculate Similarity
similarity = cosine_similarity(vectors)
```

```python
#step9.3 - Check Shape
similarity.shape
```

```python
#step10 - Create Function
def recommend(movie):
    movie = movie.lower()
    
    if movie not in new_df['title'].str.lower().values:
        print("Movie not found")
        return
    
    movie_index = new_df[new_df['title'].str.lower() == movie].index[0]
    
    distances = similarity[movie_index]
    
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    print("Recommended Movies:\n")
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("Avatar")
```

```python
recommend("Batman")
```

```python
recommend("Titanic")
```
